```python
# path = r'C:\Users\shash\Downloads\Green_Hackathon-Fall2020-main\Green_Hackathon-Fall2020-main\data\Analysis' # use your path
path = r'C:\Users\umama\shravya python assignment\Analysis'
all_files = glob.glob(path + "/*.csv")

li = []

for csv in all_files:
    cols = ['Actual','Predicted','obs_ci_lower','obs_ci_upper','Datetime']
    df = pd.read_csv(csv, index_col=None ,encoding = "ISO-8859-1",engine = 'python', sep=',', header=0,names=cols,error_bad_lines=False)
    df['Building_Name'] = os.path.basename(csv)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)



#for csv in all_files:
    #frame = pd.read_csv(csv)
    #frame['filename'] = os.path.basename(csv)
    #li.append(frame)
#frame
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

df = uncg_data

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="building_name",
        options=[{"label": x, "value": x} 
                 for x in df.columns[1:]],
        value=df.columns[1],
        clearable=False,
    ),
    dcc.Graph(id="time-series-chart"),
])

@app.callback(
    Output("time-series-chart", "figure"), 
    [Input("ticker", "value")])
def display_time_series(ticker):
    fig = px.line(df, x='date', y=ticker)
    return fig

app.run_server(debug=False)
```
